```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from sklearn.feature_selection import SelectKBest, chi2
```

```python
#load data
def load_data(path):
  df = pd.read_csv(path)
  return df


#Exploring data
def data_analysis(df):
  # Check for missing values
  df.isnull().sum()
  #data information
  print(f"data information:",df.columns)
  #data shape
  print(f"data shape:",df.shape)
  # Count the unique values in a column
  print(f"Count the different classes :",df['selection_model_y'].value_counts())
  #data visualisation
  fig = px.histogram(df,x='selection_model_y')
  fig.show()
```

```python
#analysis the Raw data
rawDataPath = "/content/gdrive/My Drive/Meta_Asr/Main_Dataset/Main_RawMetaData.csv"
df = load_data(rawDataPath)
```

```python
data_analysis(df)
```

```python
#analysis the Preprocessed data
preProcessedDataPath = "/content/gdrive/My Drive/Meta_Asr/Main_Dataset/Main_PreProcessedMetaData.csv"
df = load_data(preProcessedDataPath)
data_analysis(df)
```
